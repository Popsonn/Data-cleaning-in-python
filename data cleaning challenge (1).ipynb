```python
#cleaning the club column 
df['Club'] = df['Club'].str.replace('\n', ' ')
df['Club']
```

```python
#checking the unique columns in contract
df['Contract'].unique()
```

```python
#applying a function to extract start year
def convert_contract(x):
    if isinstance(x,str) and 'On Loan' in x:
        #x.split(',')
        start = x[-12:-8]
        return start
    else:
        return x[:4]
```

```python
#applying the function
df['Contract Start']= df['Contract'].apply(convert_contract)
```

```python
df['Contract Start']
```

```python
#extracting the end year of the contract
df['Contract End']= df['Contract'].apply(lambda x:x[-4:])
df['Contract End']
```

```python
#filling the null values with nont applicable because they werent loaned players
df['Loan Date End'].fillna('not applicable', inplace=True)
df['Loan Date End']
```

```python
#checking the unique values in height column 
df['Height'].unique()
```

```python
#seperating the feet and inches with '-'
df['Height'] = df['Height'].str.replace("\'", '-')
df['Height'].unique()
df['Height'] = df['Height'].str.replace('"','')
df['Height'].unique()
```

```python
#stripping the cm
df['Height']=df['Height'].str.strip('cm')
df['Height'].unique()
```

```python
#defining a function that converts ft and inches to cm
def convert_height(x):
    if isinstance(x,str) and '-' in x:
        feet,inches = map(int, x.split('-'))
        cm = (feet * 30.48) + (inches * 2.54)
        return round(cm)
    else:
        return (x)
#dotted_time = '1.20'
#h, m = [int(s) for s in dotted_time.split('.')]
#print h, m
#1 20
#minutes = 60*h + m
#hours = minutes / 60.0
# print minutes, hours
#80 1.33333333333
```

```python
#applying the defined function
df['Height']=df['Height'].apply(convert_height)
df['Height'].unique()
```

```python
#checking unique values in weight column
df['Weight'].unique()
```

```python
#defining a function that converts pounds to kg 
def convert_weight(x):
    if isinstance(x,str) and 'lbs' in x:
        kg=int(x[:3]) /2.205
        return float(kg)
    else:
        return int(x[:-2])
```

```python
#applying the function
df['Weight']=df['Weight'].apply(convert_weight)
```

```python
#checking that the function worked
df['Weight'].unique()
```

```python
#converting the joined column to datetime type
df['Joined']= pd.to_datetime(df['Joined'])
df['Joined']
```

```python
#checking unique values in value column
df['Value'].unique()
```

```python
#stripping the euro symbol
df['Value'] = df['Value'].str.strip('€')
df['Value']
```

```python
#defining a function that converts K to 1000 and M to 1000000
def convert_money(x):
    if  'K' in x:
        x = x.replace('K','')
        x =float(x)*1000
        return round(x)
    elif 'M' in x:
        x = x.replace('M','')
        x =float(x)*1000000
        return round(x)
    else:
         return float(x)
```

```python
#applying the function
df['Value'] = df['Value'].apply(convert_money)
```

```python
#calling the value column
df['Value']
```

```python
#checking unique values in wage column
df['Wage'].unique()
```

```python
#stripping the euro symbol
df['Wage']= df['Wage'].str.replace('€','')
df['Wage'].unique()
```

```python
#defining a function that converts K to 1000 and M to 1000000
def convert_wage(x):
    if isinstance(x,str) and 'K' in x:
        thousand =int(x[:-1]) * 1000
        return thousand
    else:
        return x
```

```python
df['Wage']=df['Wage'].apply(convert_wage)
```

```python
df['Wage'].unique()
```

```python
#cleaning the euro symbol in release clause column
df['Release Clause']= df['Release Clause'].str.replace('€','')
df['Release Clause']
```

```python
#applying the function (convert_money) since the column are similar
df['Release Clause']= df['Release Clause'].apply(convert_money)
```

```python
#calling the release clause column
df['Release Clause']
```

```python
#checking the unique values in Hits column
df['Hits'].unique()
```

```python
#filling na with 0 
df['Hits']= df['Hits'].fillna('0')
df['Hits']
```

```python
#defining a function to convert K to thousand 
def delete(x):
    if isinstance(x, str) and 'K' in x:
        x = x.strip('K')
        thousand = float(x) * 1000
        return round(thousand)
    else:
        return int(x)
    
```

```python
#applying the function
df['Hits']= df['Hits'].apply(delete)
```

```python
#checking applied function 
df['Hits'].unique()
```

```python
#drop the playerUrl and photoUrl columns
df = df.drop(columns=['playerUrl','photoUrl','Contract'])
df
```

```python
#renaming some columns such as ↓OVA,Height,Weight,Value,Wage,Release Clause
df.rename(columns = {'↓OVA':'OVA','Height': 'Height(cm)','Weight':'Weight(kg)','Value':'Value(€)','Wage':'Wage(€)','Release Clause':'Release Clause(€)'}, inplace = True)
df
```
